```python
data = loadtxt("Datos lab/Datos Ag - 8.11/Graph 1/DC-all.txt")
p0 = data[:,0]
p1 = data[:,1]
p2 = data[:,6]
p3 = data[:,7]
p4 = data[:,3]


VoltajeNorm=np.empty(0)
Angle=np.empty(0)
otro=np.empty(0)
info1=np.empty(0)
info2=np.empty(0)

for i in range(p0.size):
    if(p0[i]==2.0):
        rta=p1[i]/(p2[i]*Cte)
        VoltajeNorm=np.append(VoltajeNorm,rta)
        info1=np.append(info1,p1[i])
        info2=np.append(info2,p2[i])
        Angle=np.append(Angle,p4[i])
        otro=np.append(otro,p3[i])
        
title("normalizado 2")

plot(Angle,info1)
plot(Angle,info2)
plot(Angle,otro)

#----------------------------------------------------------------------
data = loadtxt("Datos lab/Datos Ag - 8.11/Graph 2/DC-all.txt")
p0 = data[:,0]
p1 = data[:,1]
p2 = data[:,6]
p3 = data[:,7]
p4 = data[:,3]


VoltajeNorm=np.empty(0)
Angle=np.empty(0)
otro=np.empty(0)
info1=np.empty(0)
info2=np.empty(0)

for i in range(p0.size):
    if(p0[i]==2.0):
        rta=p1[i]/(p2[i]*Cte)
        VoltajeNorm=np.append(VoltajeNorm,rta)
        info1=np.append(info1,p1[i])
        info2=np.append(info2,p2[i])
        Angle=np.append(Angle,p4[i])
        otro=np.append(otro,p3[i])
        
title("normalizado 2")

plot(Angle,info1)
plot(Angle,info2)
plot(Angle,otro)
```
