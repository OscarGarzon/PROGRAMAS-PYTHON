```python
#Se importan datos
data = np.loadtxt("KALPHA")
data2=np.loadtxt("KBETA")
U_alpha=data[:,0]
INT_alpha= data[:,1]
U_beta=data2[:,0]
INT_beta=data2[:,1]
```
