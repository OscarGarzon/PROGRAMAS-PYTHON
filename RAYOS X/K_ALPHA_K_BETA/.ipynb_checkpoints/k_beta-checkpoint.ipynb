```python
# -*- coding: utf-8 -*-
%pylab inline 
import numpy as np
%pylab --no-import-all
from matplotlib.pyplot import *
from numpy import *
import math
from scipy import stats
%matplotlib inline
figsize(7,7)
```

```python
#Se importan datos
data = np.loadtxt("KALPHA")
U=data[:,0]
INT= data[:,1]
U
```

```python
#Cálculo de incertidumbres
err_y=[]
err_x=[]
for i in range(len(U)):
    error_x=1
    err_x.append(error_x)
```

```python
m, b, r_value, p_value, std_err = stats.linregress(U,INT)
m,b,r_value**2,  std_err
```

```python
fig, ax = subplots()
ylabel('Intensidad $Imp/s$', fontsize=20)
xlabel('$V(KV$)', fontsize=20)
for i in range(len(U)):
    errorbar(U[i], INT[i], xerr=err_x[i], yerr=1,capsize=15,color="r",fmt='o')
x=arange(0.03, 140,0.001)
y=x*m+b
ax.text(0.025, 0.2, r"$y=0.5579
x+0.0335$", fontsize=18, color="black")
ax.text(0.025, 0.18, r"$r^2=0.98$", fontsize=18, color="black")

plot(x, y, color='k')
ax.legend(loc=2);
ax.legend(loc=2);
```
