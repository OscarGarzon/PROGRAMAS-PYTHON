```python
# -*- coding: utf-8 -*-
%pylab inline 
import numpy as np
%pylab --no-import-all
from matplotlib.pyplot import *
from numpy import *
import pylab as pl
import math
from scipy import stats
%matplotlib inline
figsize(7,7)
```

```python
#Se importan datos
data = np.loadtxt("caracteristica_cu")
THETA=data[:,0]
INT= data[:,1]
```

```python
#GRAFICA

fig, ax = subplots()
ylabel('Log(Intensidad)', fontsize=20)
xlabel('$\Theta$ (Grados)', fontsize=20)
pl.yticks([3,4,5,6,7,8,9],
          [r'$3$', r'$4$', r'$5$',r'$6$',r'$7$',r'$8$',r'$9$'],fontsize=20)
pl.xticks([5.5,10,20,30,40,50],[r'$5.5$',r'$10$',r'$20$',r'$30$',r'$40$',r'$50$'], fontsize=20)

plot(THETA,log(INT))
ax.legend(loc=2);
ax.legend(loc=2);
```
